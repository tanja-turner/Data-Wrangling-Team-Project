```python
# column year some of the years saved as string so I changed the type in next step
imdb_movies["year"].unique()
```

```python
def convert_str_to_int(column):
    if pd.isnull(column) == True:
        return column
    if type(column) != str:
        return column
    else:
        string = column.split(' ')[-1]
        string = string.replace("'", '')
        #string = string.replace(',', '')
        return int(string)
    
imdb_movies["year"] = imdb_movies["year"].apply(convert_str_to_int)
imdb_movies["year"].unique()
```
